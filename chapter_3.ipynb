## 3.1 Python 설치 및 실행

1. Google에 `install python` 검색 \


2. 가장 최신 버전이 아닌, 이전 버전을 선택 \


3. Python 실행
   1. Python을 실행하기 위해 Windows에서 `시작` 탭에서 `터미널` 앱을 사용할 수 있다. \
   
   2. `터미널` 앱을 키고 다음과 같이 python을 실행시킬 수 있다.
       ```bash
       PS C:\Users\dj.yoon> python
       Python 3.12.4 | packaged by Anaconda, Inc. | (main, Jun 18 2024, 15:03:56) [MSC v.1929 64 bit (AMD64)] on win32
       Type "help", "copyright", "credits" or "license" for more information.
       >>>
       ```

4. Hello, world! \
`Hello, world!` 라는 문자열을 출력해보자.
    ```python
    >>> print("Hello, world!")
    Hello, world!
    >>>
    ```
