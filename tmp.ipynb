## 3.3 가상환경 설치

### 1. 가상환경
Python을 사용하다보면 다양한 패키지를 설치하게 되는데 이들은 서로 **호환성**을 가지고 있다. \
따라서, 프로젝트를 새로 생성할 땐 **독립적으로** 패키지를 관리하는 것이 일반적이다. \
특히, [가상환경(virtual environment)](https://docs.python.org/ko/3/tutorial/venv.html)을 사용하는 방법이 권장된다.

### 2. venv
Python에서 가상환경을 만들어주는 다양한 프로그램들이 있는데 여기에선 가장 기본적인 `venv` package를 사용한다.
